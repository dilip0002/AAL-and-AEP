```python
import os
import pathlib as pl

from heatmap_summary import *
%matplotlib inline
```

```python
root_dir = pl.Path(os.getcwd())
```

```python
df = pd.read_csv(root_dir.parent/'outputs/AAL_csv_crs.csv')

threshold = 3000
book = 'Turkey'
project = 'AAL'
#DATAFIELDS = ['dmg_val_str', 'dmg_val_con', 'dmg_total']
gdf = df_to_gdf(df, book, project)
```

```python
anom_threshold = 3000
anoms = gdf[gdf['dmg_total'] > anom_threshold].copy()
get_outlier_info(gdf, anom_threshold)  
```

```python
bds = anoms.total_bounds
location = [round(bds[[1,3]].mean(),2), round(bds[[0,2]].mean(),2)]
```

```python
col, threshold = 'dmg_val_co', 10
anomdata = list(zip(anoms.geometry.y, anoms.geometry.x))
m = map_aals(anoms[anoms[col] > threshold].copy() , col, 'Contents', location=location, zoom_start=5)

for point in range(0, len(anomdata)):
    if int(anoms['dmg_val_co'].iloc[point]) > 3000:  
        folium.Marker(anomdata[point], popup='${:,}'.format(int(anoms['dmg_total'].iloc[point]))).add_to(m)
    
m
```

```python
col, threshold = 'dmg_val_st', 10
anomdata = list(zip(anoms.geometry.y, anoms.geometry.x))
m = map_aals(anoms[anoms[col] > threshold].copy() , col, 'Structure', location=location, zoom_start=5)

for point in range(0, len(anomdata)):
    if int(anoms['dmg_val_st'].iloc[point]) > 3000:  
        folium.Marker(anomdata[point], popup='${:,}'.format(int(anoms['dmg_val_st'].iloc[point]))).add_to(m)
    
m
```
